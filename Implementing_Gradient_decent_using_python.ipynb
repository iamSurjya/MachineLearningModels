```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#creating the function and plotting it 

```python
function=lambda x:((x+5)**2)
```

```python
#Get 1000 evenly spaced numbers between -1 and 3 (arbitratil chosen to ensure steep curve)
x = np.linspace(-10,10,500)

#Plot the curve
plt.plot(x, function(x))
```

#find derivative of function

```python
def deriv(x):
    x_deriv=2*(x+5)
    return x_deriv
```

```python
x_list,y_list=[],[]

def step(x_new, x_prev, precision, learning_rate):
    # keep looping until your desired precision
    while abs(x_new - x_prev) > precision:
        x_prev = x_new
        d_x = deriv(x_prev)
        x_new = x_prev - (learning_rate * d_x)
        #Storing the value of each step in list
        x_list.append(x_new)
        y_list.append(function(x_new))
    return x_new,x_list,y_list
```

```python
x_axis,y_axis=[],[]
final_x=0

final_x,x_axis,y_axis=step(3,0,0.001,0.01)
```
